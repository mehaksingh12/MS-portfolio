Linear Regression

```r
x <- c(151, 174, 138, 186, 128, 136, 179, 163, 152, 131)
y <- c(63, 81, 56, 91, 47, 57, 76, 72, 62, 48)

# Apply the lm() function.
relation <- lm(y~x)

print(relation)
```

```r
a <- data.frame(x = 170)
result <-  predict(relation,a)
print(result)
```

Multiple Regression

```r
input = mtcars[,c("mpg","disp","hp","wt")]
print(head(input))
```

```r
model <- lm(mpg~disp+hp+wt, data = input)
print(model)
```

```r
a <- coef(model)[1] #intercept
Xdisp <- coef(model)[2]
Xhp <- coef(model)[3]
Xwt <- coef(model)[4]

print(a) #intercept
print(Xdisp)
print(Xhp)
print(Xwt)
```

```r
#y = a+b1x1+b2x2+..
b1 = 221
b2 =102
b3 = 2.91
Y1=a+Xdisp*b1+Xhp*b2+Xwt*b3
print(Y1)
```

Consider annual rainfall data at a place from january 2012. Create an R time series object for a period of 12 months and plot it


```r
rainfall = c(799,1174.8,865,1334.6,635.5, 998.6,784.2,882.8,1071,774,896,1120)
#convert it to a time series
rainfall.timeseries = ts(rainfall, start=c(2024,1), frequency=12)
print(rainfall.timeseries)
```

```r
plot(rainfall.timeseries)
```

```r
rainfall1 = c(799,1174.8,865,1334.6,635.5, 998.6,784.2,882.8,1071,774,896,1120)
rainfall2 = c(689,1374.8,1365,1034.6,485.5, 598.6,824.2,882.8,1271,799,1106,1320)
combined.rainfall = matrix(c(rainfall1,rainfall2),nrow=12) #convert them to a matrix
rainfallcombined.timeseries = ts(combined.rainfall, start = c(2024,1), frequency=12)
print(rainfallcombined.timeseries)
```

```r
plot(rainfallcombined.timeseries, main="multiple time series")
```

Q2. Look at the data VADeaths, create a bar chart of these data, since you have 5 rows give 5 colours to help pick out categories, add legend to the plot

```r
VADeaths
```

```r
# Load the dataset
data("VADeaths")

# Define colors for each row (age group)
colors <- c("navy", "pink", "yellow", "gray", "red")

# Create the bar plot
barplot(
  VADeaths,
  beside = TRUE, #Display bars side by side
  col = colors,   # Assign colors
  main = "VA Death Rates by Age Group and Location",
  xlab = "Location",
  ylab = "Death Rate",
  legend.text = rownames(VADeaths),  # Add legend
  args.legend = list(x = "topright", bty = "n")  # Position the legend
)
```

```r
# Load the dataset
data("VADeaths")

# Define colors for each row (age group)
colors <- c("navy", "pink", "yellow", "gray", "red")

# Create the horizontal bar plot
barplot(
  VADeaths,
  beside = TRUE,  # Display bars side by side
  col = colors,   # Assign colors
  main = "VA Death Rates by Age Group and Location",
  xlab = "Death Rate",
  ylab = "Location",
  horiz = TRUE,   # Make the bars horizontal
  legend.text = rownames(VADeaths),  # Add legend
  args.legend = list(x = "bottomright", bty = "n")  # Position the legend
)
```
